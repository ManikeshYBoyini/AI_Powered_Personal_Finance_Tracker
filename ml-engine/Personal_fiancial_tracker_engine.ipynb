## Import necessory packages 

```python
import nltk
import re
import string
from textblob import TextBlob
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import string 
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def clean_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove digits
    text = re.sub(r'\d+', '', text)
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    
    # Join tokens back
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity
```

```python
sample = "I love using this finance tracker! It's so intuitive and helpful."

cleaned = clean_text(sample)
polarity, subjectivity = get_sentiment(cleaned)

print("Cleaned Text:", cleaned)
print("Sentiment Polarity:", polarity)
print("Sentiment Subjectivity:", subjectivity)
```
